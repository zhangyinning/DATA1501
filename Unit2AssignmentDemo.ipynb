```python
# load the libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# read the data into a DataFrame
mydf = pd.read_csv("SalesRecords.csv")
```

```python
print(mydf)
```

```python
y = mydf.iloc[0, 1:5]
```

```python
x = mydf.columns[1:5]
```

```python
# create a barplot
plt.bar(x, y)

plt.title("Comparison of No# of Contracts Between Sales Representative ")
plt.xlabel("Sales Representatives")
plt.ylabel("No# of Contracts in January")
plt.show()
```

```python
# data for x-axis for the line plot
x1 = mydf.iloc[:,0]
print(x1)
```

```python
y1 = mydf.iloc[:, 1:]
print(y1)
```

```python
# Create a line plot.
lg = mydf.columns[1:5]
x2 = ["Jan.", "Feb.", "Mar.", "Apr.", "May", "Jun.", "Jul.", "Aug.", "Sep.", "Oct.", "Nov.", "Dec."]
plt.plot(x2, y1, linestyle = ":", marker = 'o')

plt.legend(lg)
plt.show()
```

```python
# create a pie chart
x3 = mydf.iloc[0, 1:]
lb = mydf.columns[1:5] 
# shall add title here. Or you loose 10 points
plt.pie(x3, 
        explode = [0.02, 0.05, 0.05, 0.02], # the distance of the pie from the center
        labels = lb, # labels such as mary, jason
        autopct = "%0.2f%%" )  # it displays percentages on the pie.
plt.show()

```
